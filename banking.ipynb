```python
plt.figure(figsize=(12,8))
correlation_matrix = df[['AI_Adoption', 'Profitability', "Efficiency",'Risk_Management',"Customer_Satisfaction","Operational_Effectiveness"]].corr()
sns.heatmap(correlation_matrix,annot=True)
plt.title("correlation Matrix of AI adoption and Financial performance Metrics")
```

```python
#carry out linear regression
```

```python
import statsmodels.api as sm
```

```python
y = df['Profitability']
```

```python
X = df[['Efficiency','Risk_Management','Customer_Satisfaction','Operational_Effectiveness', 'AI_Adoption']]
```

```python
X = sm.add_constant(X)
```

```python
model = sm.OLS(y,X).fit()
```

```python
print(model.summary())
```
